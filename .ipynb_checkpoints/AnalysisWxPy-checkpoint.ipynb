```python
# Dependencies and Setup
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os



#Import called data from CSV
data = pd.read_csv("data.csv")
data.head()
```

```python
#trimming dataframe
data = data.drop(columns="Unnamed: 0")
data = data.iloc[:500]
data["date"] = pd.to_datetime(data["date"], unit='s')
data["date"] = data["date"].dt.strftime("%m/%d/%Y")
```

```python
print(data.describe())
print("")
print(data.head())
print("")
print("There are " + str(data.name.nunique()) + " unique cities in this list.")
```

```python
day = data["date"].max()
day
```

#### Latitude vs. Temperature Plot

```python
plt.close()
plt.scatter(data.lat,data.temp,c=data.temp,marker="o",edgecolors="grey")
plt.colorbar()
plt.grid()
plt.xlim(-90,90)
plt.ylim(20,105)
#Set a Title and labels
plt.title(f"City Latitude vs. Temperature ({day})")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.tight_layout
plt.savefig("lat_temp.png")
plt.show()
```

#### Latitude vs. Humidity Plot

```python
plt.close()
plt.scatter(data.lat,data.humidity, c=data.humidity,marker="o",edgecolors="grey")
plt.colorbar()
plt.grid()
plt.xlim(-90,90)
plt.ylim(0,105)
#Set a Title and labels
plt.title(f"City Latitude vs. Humidity ({day})")
plt.xlabel("Latitude")
plt.ylabel("Humidity %")
plt.tight_layout
plt.savefig("lat_humidity.png")
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
plt.close()
plt.scatter(data.lat,data.cloud,c=data.cloud,marker="o",edgecolors="grey")
plt.colorbar()
plt.grid()
plt.xlim(-90,90)
plt.ylim(0,105)
#Set a Title and labels
plt.title(f"City Latitude vs. Cloudiness ({day})")
plt.xlabel("Latitude")
plt.ylabel("Cloudy %")
plt.tight_layout
plt.savefig("lat_cloud.png")
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
plt.close()
plt.scatter(data.lat,data.wind,c=data.wind,marker="o",edgecolors="grey")
plt.colorbar()
plt.grid()
plt.xlim(-90,90)
plt.ylim(0,31)
#Set a Title and labels
plt.title(f"City Latitude vs. Wind Speed ({day})")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.tight_layout
plt.savefig("lat_wind.png")
plt.show()
```
